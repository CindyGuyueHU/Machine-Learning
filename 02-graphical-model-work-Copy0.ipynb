### Part 1a: Men vs Women

Compute the following probabilities (show details of your reasoning):
* probability that a person passes the course
* probability that a woman passes the course
* probability that a man passes the course

### Solution description

Denote the following variables:

* **G**ender with states (W, M), where W for women, M for men
* **R**esult with states (P, F), where P for passed, F for failed

The total sample number : 

\begin{align*}
    N &= 8 + 1 + 23 + 3 + 19 + 7 + 6 + 3 \\
    &= 70
\end{align*}

The joint probability:

\begin{align*}
    p(G = W, R = P) = 27 / N \\
    p(G = W, R = F) = 8 / N \\
    p(G = M, R = P) = 29 / N \\
    p(G = M, R = F) = 6 / N \\
\end{align*}

* probability that a person passes the course

    Marginalize using sum rule:

\begin{align*}
    p(R = P) &= \sum_{G} p(G, R = P)\\
    &= p(G = W, R = P) + p(G = M, R = P)\\
    &= \frac {27 + 29} {70} = 0.8
\end{align*}
    
* probability that a woman passes the course

    Condition probability using the product rule and sum rule:

\begin{align*}
    p(R = P | G = W) &= \frac {p(G = W, R = P)} {p(G = W)} \\
    &= \frac {p(G = W, R = P)} {\sum_{R} p(R, G = W)} \\
    &= \frac {p(G = W, R = P)} {p(G = W, R = P) + p(G = W, R = F)} \\
    &= \frac {27} {27 + 8} = 0.7714
\end{align*}

* probability that a man passes the course

    Condition probability using the product rule and sum rule:
    
\begin{align*}
    p(R = P | G = M) &= \frac {p(G = M, R = P)} {p(G = M)} \\
    &= \frac {p(G = M, R = P)} {\sum_{R} p(R, G = M)} \\
    &= \frac {p(G = W, R = P)} {p(G = M, R = P) + p(G = M, R = F)} \\
    &= \frac {29} {29 + 6} = 0.8286
\end{align*}

## Part 2: (6 points) Markov Chain Monte Carlo

We consider a special case of the Metropolis Hastings algorithm. 
Recall that the Metropolis Hastings algorithm uses the current sample to decide which next sample to produce.
We use the proposal distribution which is the conditional distribution of a single variable, conditioned on the current state of all the other variables. We cycle through all the variables at each time step, updating the current state. That is for each step, we sample:
* p(A | B,C,D,E,F)
* p(B | A,C,D,E,F)
* p(C | A,B,D,E,F)
* p(D | A,B,C,E,F)
* p(E | A,B,C,D,F)
* p(F | A,B,C,D,E)

After each sampling step, we use the sample as the new state of the variable.

````python
# Solution goes here

class Node:
    """ The class of variable nodes in the Bayesian network
        - name: the name of the variable
        - state_list: possible state of this variable
        - pa: list of parent nodes
        - cond_p: the conditional probability of this variable given the states of parent nodes
        - state: the current state of this variable
        - sample: the state of this variable in a list of samples
    """
    def __init__(self, name, state_list, pa, cond_p, state):
        self.name = name
        self.state_list = state_list
        self.pa = pa
        self.cond_p = cond_p
        self.state = state
        self.sample = []

        
def initial_sample(nodes):
    """ remove all the samples in a list of nodes
    input: nodes - a list of nodes
    output: nodes - cleared nodes
    """
    for i in range(len(nodes)):
        nodes[i].sample = []
    return nodes

def Metropolis_Hastings(nodes, record):
    """ implementing the Metropolis Hasting algorithm to generate a new state (a sample) from the current state of 
        all the nodes in a Bayesian network
    input: nodes - a list of nodes in a Bayesian network
           record - if the generated sample should be recorded
    output: nodes - a list of nodes with state changed in the Markov Chain, and with samples recorded (if record = Ture)
    
    """
    
    # create samples for all nodes in a one by one manner
    for i in range(len(nodes)):
        
        # v = current node
        v = nodes[i]
        pa = v.pa    
        cond_p = v.cond_p[:]
        
        # compute the conditional probability of v according to the current state of all its parent nodes
        for j in range(len(pa)):
            cond_p = np.take(cond_p, [pa[j].state_list.index(pa[j].state)], axis=j + 1)
        cond_p = cond_p.flatten()
        
        # draw samples from the conditional probability of v
        z = np.random.multinomial(1, cond_p, size=1)
        idx = np.where(z[0] == 1)[0]
        nodes[i].state = nodes[i].state_list[idx[0]]
        
        # record the sample in v if necessary
        if record == True:
            nodes[i].sample.append(nodes[i].state[:])
            
    return nodes

def Markov_chain_Monte_Carlo(nodes, T, Burn_in):
    """ Use Markov Chain Monte Carlo to create samples
    input: nodes - a list of nodes in a Bayesian network
           T - required number of samples 
           Burn_in - number of Burn_in samples
    output: nodes
    """
    
    # clear the samples in all nodes
    nodes = initial_sample(nodes)
    
    # create burn_in samples and discard them
    for t in range(Burn_in):
        nodes = Metropolis_Hastings(nodes, False)
    
    # create samples using the Metropolis Hasting algorithm
    for t in range(T):
        nodes = Metropolis_Hastings(nodes, True)
    
    return nodes
````

### Part 2b: Proposed algorithm is a valid MCMC method

In the lectures, you have seen that Metropolis Hastings algorithms need to satisfy certain properties. State these properties and show that the proposed algorithm satisfies them.

### Solution description

In this special case of the Metropolis Hastings algorithm,

####1.Initialize 

$$
    p(A^{(0)}, B^{(0)}, C^{(0)}, D^{(0)}, E^{(0)}, F^{(0)}) = p^{*}(A=False, B=none, C=False, D=healthy, E=False, F=False)
$$


####2.Assuming T is the total iteration. For $\tau = 1, ..., T$, we in turns:
  
  replacing the value of one of the variables by a value drawn from the conditional distribution of that variable given all other variables

  * Step A: Sample $A^{(\tau + 1)} \sim p^{*}(A | B^{(\tau)})$
  * Step B: Sample $B^{(\tau + 1)} \sim p^{*}(B)$
  * Step C: Sample $C^{(\tau + 1)} \sim p^{*}(C)$
  * Step D: Sample $D^{(\tau + 1)} \sim p^{*}(D | B^{(\tau + 1)}, C^{(\tau + 1)})$
  * Step E: Sample $E^{(\tau + 1)} \sim p^(E | D^{(\tau + 1)})$
  * Step F: Sample $F^{(\tau + 1)} \sim p^(F | C^{(\tau + 1)})$
  
To show this procedure samples from the required distribution $p^{*}(A,B,C,D,E,F)$, we need to prove it satisfies the following two properties:

** Property 1: The required distribution $p^{*}(A,B,C,D,E,F) $ is invariant of the Markov chain**

According to the Bayesian network given in the question, we have the joint probability:

$$
    p^{*}(A,B,C,D,E,F) = p^{*}(A|B) p^{*}(B)  p^{*}(C) p^{*}(D|B,C)  p^{*}(E|D)  p^{*}(F|C) \tag{1}
$$

we prove that $p^{*}(A,B,C,D,E,F)$ is invariant of each step in 2., hand hence is invariant of the whole Markov chain:

For example, assume given $(A^{(\tau)}, B^{(\tau)}, C^{(\tau)}, D^{(\tau)}, E^{(\tau)}, F^{(\tau)}) \sim p^{*}(A,B,C,D,E,F)$. We prove $p^{*}$ is invariant after **step A**, by proving the **detailed balance**, defined by:

$$
    p^{*}(\mathbf z)T(\mathbf z, \mathbf z^{'}) = p^{*}(\mathbf z^{'}) T(\mathbf z^{'}, \mathbf z) \tag{2}
$$

where  $\mathbf z$ is the state $(A^{(\tau)}, B^{(\tau)}, C^{(\tau)}, D^{(\tau)}, E^{(\tau)}, F^{(\tau)})$, $\mathbf z^{'}$ is the state after step A, $(A^{(\tau + 1)}, B^{(\tau)}, C^{(\tau)}, D^{(\tau)}, E^{(\tau)}, F^{(\tau)})$

According to the definition of $p^{*}$ in (1):

\begin{align*}
    p^{*}(\mathbf z) &= p^{*}(A, B, C, D, E, F) \\
    &= p^{*}(A|B) p^{*}(B)  p^{*}(C) p^{*}(D|B,C)  p^{*}(E|D)  p^{*}(F|C) \tag{3}
\end{align*}

\begin{align*}
    p^{*}(\mathbf z^{'}) &= p^{*}(A^{'}, B^{'}, C^{'}, D^{'}, E^{'}, F^{'}) \\
    &= p^{*}(A^{'}|B^{'}) p^{*}(B^{'})  p^{*}(C^{'}) p^{*}(D^{'}|B^{'},C^{'})  p^{*}(E^{'}|D^{'})  p^{*}(F^{'}|C^{'}) 
\end{align*}

Because $B,C,D,E,F$ are unchanged,

$B^{'} = B, C^{'} = C, D^{'} = D, E^{'} = E, F^{'} = F$ in the above equation, we then have,

$$
     p^{*}(\mathbf z^{'}) = p^{*}(A^{'}|B) p^{*}(B)  p^{*}(C) p^{*}(D|B,C)  p^{*}(E|D)  p^{*}(F|C) \tag{4}
$$

Accordint to the definition of Step A:

\begin{align*}
    T(\mathbf z, \mathbf z^{'}) &= q(\mathbf z^{'} | \mathbf z) \\
    &= q(A^{'}, B^{'}, C^{'}, D^{'}, E^{'}, F^{'} | A, B, C, D, E, F) \\
    &= p^{*}(A^{'}| B) \tag{5}
\end{align*}

Similarly,

\begin{align*}
    T(\mathbf z^{'}, \mathbf z) &= q(\mathbf z | \mathbf z^{'}) \\
    &= q( A, B, C, D, E, F | A^{'}, B^{'}, C^{'}, D^{'}, E^{'}, F^{'}) \\
    &= p^{*}(A| B^{'}) \tag{6}
\end{align*}

Substuting (3), (4), (5), (6) into (2),

\begin{align*}
    p^{*}(\mathbf z)T(\mathbf z, \mathbf z^{'}) &= (p^{*}(A|B) p^{*}(B)  p^{*}(C) p^{*}(D|B,C)  p^{*}(E|D)  p^{*}(F|C)) p^{*}(A^{'}| B) \\
    &= (p^{*}(A^{'}|B) p^{*}(B)  p^{*}(C) p^{*}(D|B,C)  p^{*}(E|D)  p^{*}(F|C))p^{*}(A| B^{'}) \\
    &= p^{*}(\mathbf z^{'}) T(\mathbf z^{'}, \mathbf z)
\end{align*}

as required.

We have proved $p^{∗}$ is invariant after step A. Similar, after step B, step C, step D, step E, and step F, $p^{∗}$ is invariant. 

Hence, $p^{*}(A,B,C,D,E,F)$ is invariant of the whole Markov chain.

** Property 2: Ergodicty: for $m \rightarrow \infty$, the distribution $p(A^{(m)}, B^{(m)}, C^{(m)}, D^{(m)}, E^{(m)}, F^{(m)})$ converges to the required invariant distribution $p^{*}(A^{(m)}, B^{(m)}, C^{(m)}, D^{(m)}, E^{(m)}, F^{(m)})$, irrespective of the choice of initial distribution $p(A^{(0)}, B^{(0)}, C^{(0)}, D^{(0)}, E^{(0)}, F^{(0)})$**

Because none of the conditional distributions is anywhere zero,

any point in the $\mathbf z$ space can be reached from any other point in a finite number of steps.

According to the Ergodic theorem, the Markov Chain is ergodic.

In property 1, we have proved that $p^{*}(A,B,C,D,E,F)$ is a equilibrium distribution of the Markov Chain. Because one ergodic Markov Chain can have only one equilibrium distribution, the distribution of samples drawn from the Markov Chain will converge to $p^{*}$ in finite steps.

Notice that, there is another important property of this special case of the Metropolis Hastings algorithm:

** Property 3: The acceptance probability $A(\mathbf z^{*}|\mathbf z) = 1$, which means all the candidate samples are accepted.**

Consider a single step in 2., becauseit only involving in one variable $z_k$ and other variables remaining the same,

the transition probability from $\mathbf z$ to $\mathbf z^{*}$ is given by

$$
    q_k(\mathbf z^{*}| \mathbf z) = p^{*}(z_k^{*} | \mathbf z_{ \backslash k}) \tag{7}
$$

Note that

$$
    \mathbf z^{*}_{\backslash k} = \mathbf z_{\backslash k} \tag{8}
$$

because these variables are staying the same.

Also,

$$
    p(\mathbf z) = p(z_k|\mathbf z_{\backslash k})p(\mathbf z_{\backslash k}) \tag{9}
$$

Substituting (3), (4) and (5) into the fomular for $A(\mathbf z^{*}|\mathbf z)$,

\begin{align*}
    A(\mathbf z^{*}|\mathbf z) &= \frac{p(\mathbf z^{*})q_k(\mathbf z | \mathbf z^{*})}{p(\mathbf z)q_k(\mathbf z^{*}| \mathbf z)} \\
    &= \frac {p(z_k^{*}|\mathbf z^{*}_{\backslash k}) p (\mathbf z^{*}_{\backslash k}) p(z_k|\mathbf z^{*}_{\backslash k})} {p(z_k | \mathbf z_{\backslash k})p(\mathbf z_{\backslash k})p(z^{*}_k | \mathbf z_{\backslash k})} \\
    &= 1
\end{align*}

which means the candidate sample generated by the Metropolis-Hasting steps are always accepted.

###Solution Description

#### Construct Factors

Assuming the Factors presented to us are 2-Dimensional tables (as the probability tables), our goal is reshape the Factors to N-Dimensional matrix, where N = the number of variables of this factor.

In the ith dimension of the factor matrix, there are $d_i$ possibilities, where $d_i$ equals the number of the possible values of descrite variable $V_i$, $i = 1,.., N$

Here we assume that for each factor, the variables and their order are known.

#### Factor Graphs

The factor graph information are stored in two types of nodes: Variable nodes and Factor nodes.

* Variables to Factors connection

f_list: which stores all the factors connected to a variable
    
* Factors to Variables connection

v_list: which stores all the variables connected to a factor

#### Message passing

* Message sent by the leaf nodes:

$$
    \mu_{x \rightarrow f} (x) = 1 \\
    \mu_{f \rightarrow x} (x) = f(x)
$$

* Message sent by a variable node to an adjacent factor node:

$$
    \mu_{x_m \rightarrow f_s} (x_m) = \prod_{l \in ne(x_m) \backslash f_s} \mu_{f_l \rightarrow x_m} (x_m)
$$

Because each of the $\mu_{f_l \rightarrow x_m} (x_m)$ on the right hand side are a function of $x_m$, for discrete we $x_m$ can do elementwise product on them to get the left hand side $\mu_{x_m \rightarrow f_s} (x_m)$, which is also a function of $x_m$.

* Message sent by a factor node to an adjacent variable node:

$$
    \mu_{f_s \rightarrow x} (x) = \sum_{x_1,.., x_M}f_s(x, x_1, ..., x_M) \prod_{m \in ne(f_s) \backslash x} \mu_{x_m \rightarrow f_s} (x_m)
$$

suppose $M$ is the list of $ne(f_s)$, $M_1$ is the sub-list contains all the element in the left hand side of $x$ in $M$, $M_2$ is the sub-list contains all the element in the right hand side of $x$ in $M$

we have 
$$ 
    \mu_{f_s \rightarrow x} (x) = \sum_{x_1,.., x_M} \prod_{m \in M_1} \mu_{x_m \rightarrow f_s} (x_m) f_s(x, x_1, ..., x_M) \prod_{m \in M_2} \mu_{x_m \rightarrow f_s} (x_m) \tag{1}
$$

In this implementation, we can compute the sum and product in one step as shown below:

In the case that there is only one element in $M$, say $\mu_{x_1 \rightarrow f_s} (x_1)$, $f_s(x, x_1) \in \RR^{d_0 \times d_1}$

$$
    \mu_{f_s \rightarrow x} (x) = \sum_{x_1}f_s(x, x_1) \mu_{x_1 \rightarrow f_s} (x_1) \tag{2}
$$

Let $\mathbf F$ be a 2-dimensional matrix presenting the values of $f_s(x, x_1)$, $\mathbf m$ be the vector of the message $\mu_{x_1 \rightarrow f_s} (x_1)$, $\mathbf F \in \RR^{d_0 \times d_1}$,  $\mathbf m \in \RR^{d_1}$, the product

$$
    \mathbf F \mathbf m =   \left[ \begin{array}{ccc}
    \mathbf F_{11} \mathbf m_1 + \mathbf F_{12}\mathbf m_2 + ... + \mathbf F_{1d_1}\mathbf m_{d_1} \\
    \mathbf F_{21} \mathbf m_1 + \mathbf F_{22}\mathbf m_2 + ... + \mathbf F_{2d_1}\mathbf m_{d_1} \\
    ...\\
    \mathbf F_{d_01}\mathbf m_1 + \mathbf F_{d_02}\mathbf m_2 + ... + \mathbf F_{d_0d_1}\mathbf m_{d_1} \end{array} \right]
$$

gives a vector in $\RR^{d_0}$ presenting the message $\mu_{f_s \rightarrow x} (x)$ in (2)

This conclusion can be generalized to (1). the matrix of $f_s(x, x_1, ..., x_M)$ has $|M|+1$ dimensions as discussed above. 

Because the matrices of factors are correctly reshaped, when we product a $D$-dimensional matrix by a vector, we get a $(D-1)$-dimensional matrix. By repeating this operation over all the variables $m \in M$, we eliminate the dimensions one by one. 

As there are $|M|$ variables in $M$, we finally get a one dimensional vector in $\RR^{d_0}$ presenting the message $\mu_{f_s \rightarrow x} (x)$, where $d_0$ is the possible values of $x$.

#### Compute marginal distribution

$$
    p(x) = \prod_{s \in ne(x)} \mu_{f_s \rightarrow x} (x)
$$

#### Recording the calculated messages

We record the calculated message, as a hash map from the destination node to the mu, in the node sending it.

So, for the same graphical model, one message can be calculated at most once. (When we need the message in the second time, we can read it directly from the two tables.)

In pratice, this method has the same efficiency as the method of  "propagate from leafs to root, and than from the root to leafs" when required to find marginal probabilities for all variables, but more efficient for marginalizing for only one variable.

Notice that, 

because the Bayesian network is already correctly marginalized, we don't worry about normalizing the result in the sum-product algorithm

````python
def factor_to_variable_mu(f, v):
    """Message sent by a factor node f, to an adjacent variable node v
    input - f: the factor node
            v: the variable node
    output - mu: the message from f to v
    """
    
    # get the list of all adjacent variables of f, and the position of v in the list
    lis = f.v_list[:]
    idx = lis.index(v)
    
    # if f is a leaf node, return the message sent by a factor leaf node
    # if f is not a leaf node, return F[f]
    mu = initial_factor_leaf_mu(f)
    
    # for all the variable nodes v_i in the left hand side of v in lis
    # use the message sent from v_i to f to eliminate one dimension from mu
    i = len(lis)-1
    while i > idx:
        if f in lis[i].mu.keys():
            mu = np.dot(mu, lis[i].mu.get(f))
        else:
            mu = np.dot(mu, variable_to_factor_mu(lis[i], f))
        i -= 1
    
    # for all the variable nodes v_i in the right hand side of v in lis
    # use the message sent from v_i to f to eliminate one dimension from mu
    mu = mu.T
    for i in range(idx):
        if f in lis[i].mu.keys():
            mu = np.dot(mu, lis[i].mu.get(f))
        else:
            mu = np.dot(mu, variable_to_factor_mu(lis[i], f))
    
    # if the message from f to v is not recorded in the F_to_V_mu_table, record it
    if not v in f.mu.keys():
        f.mu[v] = mu
        
    return mu


def variable_to_factor_mu(v, f):
    """Message sent by a variable node v, to an adjacent factor node f
    input - v: the variable node
            f: the factor node
    output - mu: the message from v to f
    """
    
    # get the list of all adjacent factors of v, and delete f from the list
    lis = v.f_list[:]
    del lis[lis.index(f)]
    
    # if f is a leaf node, return the message sent by a factor leaf node
    # if f is not a leaf node, return a vector of 1
    mu = initial_variable_leaf_mu(v)
    
    # for all the factors in lis f_i, perform elementwise product on (the message from f_i to v) and (mu)
    for i in range (0, len(lis)):
        if v in lis[i].mu.keys():
            mu = mu * lis[i].mu.get(v)
        else:
            mu = mu * factor_to_variable_mu(lis[i], v)
    
    # if the message from v to f is not recorded in the V_to_F_mu_table, record it
    if not f in v.mu.keys():
        v.mu[f] = mu
    
    return mu
````

````python
# solution goes here

class Variable:
    """ The class of variable nodes
        - name: the name of the variable
        - dim: the possible states of the variable
        - f_list: all factors adjacent to the variable
        - mu: a hash map of all messages sent by the variable
    """
    def __init__(self, name, dim):
        self.name = name
        self.dim = dim
        self.f_list = []
        self.mu = {}
    def connect(self, f):
        """ connect f to the variable"""
        self.f_list.append(f)
    def disconnect(self, f):
        """ disconnect f from the variable"""
        del self.f_list[self.f_list.index(f)]

class Factor:
    """ The class of factor nodes
        - name: the name of the factor
        - F: the N dimensional array of this factor (N = the number of variables connected to this factor)
        - v_list: all variables adjacent to the variable
        - mu: a hash map of all messages sent by the factor
    """
    def __init__(self, name, F):
        self.name = name
        self.F = F
        self.v_list = []
        self.mu = {}
    def connect(self, v):
        """ connect v to the factor"""
        self.v_list.append(v)
    def disconnect(self, v):
        """ disconnect v from the factor"""
        del self.v_list[self.v_list.index(v)]

def connect_nodes(lis):
    """ connect nodes(batch)
    input - lis: a list of nodes to be connected
    """
    for i in range(len(lis)):
        node = lis[i][0]
        adjacent_nodes = lis[i][1]
        for j in range(len(adjacent_nodes)):
            node.connect(adjacent_nodes[j])

# Create the Variable instances
A = Variable("A", 2)
B = Variable("B", 3)
C = Variable("C", 2)
D = Variable("D", 4)
E = Variable("E", 2)
F = Variable("F", 2)

# Create the Factor instances
f_AB = Factor("f_AB", p_AgB * p_B)
f_BCD = Factor("f_BCD", np.reshape(p_DgBC, (D.dim, C.dim, B.dim), order = "C"))
f_C = Factor("f_C", p_C)
f_DE = Factor("f_DE", p_EgD)
f_CF = Factor("f_CF", p_FgC)

# connect the Factors to Variables (batch)
connect_nodes([[f_AB, [A, B]], 
               [f_BCD, [D, C, B]], 
               [f_C, [C]], 
               [f_DE, [E, D]],
               [f_CF, [F, C]]])

# connect the Variables to Factors (batch)
connect_nodes([[A, [f_AB]],
               [B, [f_AB, f_BCD]],
               [C, [f_BCD, f_C, f_CF]],
               [D, [f_BCD, f_DE]],
               [E, [f_DE]],
               [F, [f_CF]]])

def sum_product(variable_list, values):
    """Implement the sum-product algorithm to calculate the marginal probability for a given list of variables
    input - variable_list: the list of variables we want to calculate marginal probability for
            values: discrete values of variables in variable_list
    output - P: the marginal probability
    """
    # for all variables v in variable_list
    for i in range (len(variable_list)):
        
        # compute the marginal probability as the elementwise production of all messages from adjacent factors to v
        v = variable_list[i]
        P = np.ones(v.dim)
        for j in range(0, len(v.f_list)):
            P = P * factor_to_variable_mu(v.f_list[j], v)
        
        # print the marginal probability
        value_list = values[i]
        for j in range(len(value_list)):
            print "P(",value_list[j],") = ", P[j]
        
    return P

# compute p(E)
P_E = sum_product([E],[["E = False", "E = True"]])
````

## Part 4: (3 points) Interpreting results

Given that you have a cough (i.e. $C=True$), should you go to the hospital? Justify your decision by comparing the probability of an emergency versus the probability of your cough being fast healing.

### Solution description

According to the sum rule:

$$
    p(E|C=True) = \sum_{A,B,D,F} p(A,B,D,E,F|C = True) \tag{1}
$$

Because in the Bayesian network, C has no parent nodes, according to do product rule we can write:
    
$$
    p(A,B,C,D,E,F) = p(A,B,D,E,F|C)p(C) \tag{2}
$$

We set $p'(C=True) = 1$ and $p'(C=False) = 0$ in (2), we can obtain

\begin{align*}
    p'(A,B,C = True,D,E,F) &= p(A,B,D,E,F|C = True)p'(C = True)\\
    &= p(A,B,D,E,F|C = True) \tag{3}
\end{align*}

while,

\begin{align*}
    p'(A,B,C = False,D,E,F) &= p(A,B,D,E,F|C = Flase)p'(C = False) \\
    &= 0 \tag{4}
\end{align*}

Combining (3) and (4) with (1), 

\begin{align*}
    p(E|C=True) &= \sum_{A,B,D,F} p(A,B,D,E,F|C = True) \\
    &= \sum_{A,B,D,F} p'(A,B,C = True,D,E,F) \\
    &= \sum_{A,B,D,F} (p'(A,B,C = True,D,E,F) + p'(A,B,C = False,D,E,F)) \\
    &= \sum_{A,B,C,D,F} p'(A,B,C,D,E,F)
\end{align*}

Which give the solution to the question:

* To obtain a conditional probability, say $p(E|C = True)$, we simply replace the prior probability $p(C)$ by $p'(C)$ in the Bayesian network, and marginalize out all the other variables in the new joint probability $p'(A,B,C,D,E,F)$ using the sum-product rule, which gives the solution  $p'(E) = p(E|C = True)$

Given $C = True$,

The probability of an emergency: $P(E = True|C = True) = 0.10$

The probability of the cough being fast healing: $P(F = True | C = True) = 0.80$

By comparing these two probabilities, we can conclude that without other evidences: it is much more likely that a cough being fast heal, than it becoming a emergency. 

**Simply by comparing the conditional probability, if you only have a cough, you should not go to the hospital until other symptoms appear.**


Notice that, another consideration should be the losses if you make a wrong prediction. For example,

if you predict $(F = False)$ has a higher probability given $(C = True)$, and you go to the hospital. The result turns out $(F = true)$ (the cough is a fast healing), the only things you can lose may be the time and money for some medical diagnosis. 

However, if you predict $(E = False)$ has a higher probability given $(C = True)$, and you don't go to the hospital. The result turns out $(E = true)$ (the cough becomes emergency), you may lose your life.

The losses of wrong prediction depends on how you define the *lost functions*.

**In practice, to making decision in pratice, we should consider both the conditional probability and the lost functions.**
