## (3 points) Classification via Fisher's Linear Discriminant

Consider the problem of binary classification. Fisher's criterion is given by:
$$
J(w) = \frac{w^T S_B w}{w^T S_W w}
$$
where $S_B$ is the between class covariance and $S_W$ is the within class covariance.

When implementing this as a linear classifier, you need to choose a threshold in projection space. Describe your criteria for choosing a threshold, and justify why it is a good one.

### Solution description


Implement two functions ```train_fld``` and ```predict_fld``` that corresponds to training and prediction. The functions should be used as in the following cell.

#### 1. prove the matrix form of computing $S_w$
According to definition
$$
    S_w = \sum_{n \in \mathcal {C}_1}(\mathbf{x}_n - \mathbf{m}_1)(\mathbf{x}_n - \mathbf{m}_1)^T + \sum_{n \in \mathcal {C}_2}(\mathbf{x}_n - \mathbf{m}_2)(\mathbf{x}_n - \mathbf{m}_2)^T \tag{1}
$$

Where $\mathbf{x}_n , \mathbf{m}_1 , \mathbf{m}_2 \in \RR^{D \times 1}$

Let $ \mathbf{X}_1 = \left[ \begin{array}{ccc}
\mathbf{x}_1^T \\
\mathbf{x}_2^T \\
...\\
\mathbf{x}_n^T \end{array} \right] $ for $\mathbf{x}_n \in \mathcal{C}_1$
, $ \mathbf{X}_2 = \left[ \begin{array}{ccc}
\mathbf{x}_1^T \\
\mathbf{x}_2^T \\
...\\
\mathbf{x}_n^T \end{array} \right] $ for $\mathbf{x}_n \in \mathcal{C}_2$
, $ \mathbf{M}_1 = \left[ \begin{array}{ccc}
\mathbf{m}_1^T \\
\mathbf{m}_1^T \\
...\\
\mathbf{m}_1^T \end{array} \right] $
, $ \mathbf{M}_2 = \left[ \begin{array}{ccc}
\mathbf{m}_2^T \\
\mathbf{m}_2^T \\
...\\
\mathbf{m}_2^T \end{array} \right] $
 
Then $\mathbf{X}_1 - \mathbf{M}_1 =  \left[ \begin{array}{ccc}
(\mathbf{x}_1 - \mathbf{m}_1)^T \\
(\mathbf{x}_2 - \mathbf{m}_1)^T \\
...\\
(\mathbf{x}_n - \mathbf{m}_1)^T \end{array} \right]$
, $(\mathbf{X}_1 - \mathbf{M}_1)^T =  \left[ \begin{array}{ccc}
(\mathbf{x}_1 - \mathbf{m}_1) 
(\mathbf{x}_2 - \mathbf{m}_1) 
...
(\mathbf{x}_n - \mathbf{m}_1) \end{array} \right]$ 

for $\mathbf{x}_n \in \mathcal{C}_1$

Hence we have
$$
    \sum_{n \in \mathcal {C}_1}(\mathbf{x}_n - \mathbf{m}_1)(\mathbf{x}_n - \mathbf{m}_1)^T = (\mathbf{X}_1 - \mathbf{M}_1)^T (\mathbf{X}_1 - \mathbf{M}_1) \tag{2}
$$

Similarly,
$$
    \sum_{n \in \mathcal {C}_2}(\mathbf{x}_n - \mathbf{m}_2)(\mathbf{x}_n - \mathbf{m}_2)^T = (\mathbf{X}_2 - \mathbf{M}_2)^T (\mathbf{X}_2 - \mathbf{M}_2) \tag{3}
$$

Substitute (2) and (3) into (1), we get

$$
    S_w = (\mathbf{X}_1 - \mathbf{M}_1)^T (\mathbf{X}_1 - \mathbf{M}_1) + (\mathbf{X}_2 - \mathbf{M}_2)^T (\mathbf{X}_2 - \mathbf{M}_2)
$$


#### 2. Deciding the threshold c
Assume the median age in a region randomly distributed, which is a resonable assumption according to the nature of the data in this question. The best decision boundary(threshold c) should be weighted average of $m_1$ and $m_2$:

$$
    c = \frac{N_1 m_1 + N_2 m_2}{N_1 + N_2} \tag{4}
$$

where $m_1 = \mathbf{w}^T \mathbf{m}_1$, $m_2 = \mathbf{w}^T \mathbf{m}_2$, $N_1$ is the number of input data points from $\mathcal{C}_1$, $N_2$ is the number of input data points from $\mathcal{C}_2$, and

$$ 
    \mathbf{m}_1 = \frac{1}{N_1} \sum_{n \in \mathcal{C}_1} \mathbf {x}_n \tag{5}     
$$
$$
    \mathbf{m}_2 = \frac{1}{N_2} \sum_{n \in \mathcal{C}_2} \mathbf {x}_n \tag{6}
$$

Combining (5) and (6), we have

$$
    N_1 \mathbf{m}_1 + N_2 \mathbf{m}_2 = \sum_{n = 1}^{N} \mathbf{x}_n
$$

where $N = N1 + N2$ is the total numbe of input data points

Because our input data is nomalized, we have

$$
    \sum_{n = 1}^{N} \mathbf{x}_n = \mathbf{0}
$$

which gives

$$
    N_1 \mathbf{m}_1 + N_2 \mathbf{m}_2 = \mathbf{0}
$$

Thus we have

\begin{align*}
    N_1 m_1 + N_2 m_2 &= N_1 \mathbf{w}^T \mathbf{m}_1 + N_2 \mathbf{w}^T \mathbf{m}_2\\
    &= \mathbf{w}^T (N_1 \mathbf{m}_1 + N_2 \mathbf{m}_2) \\
    &= \mathbf{w}^T \mathbf{0} \\
    &= 0 \tag{7}\\
\end{align*}

Substitute (7) into (4), we have
$$
    c = 0 \tag{8}
$$

which gives a resonable of choice on threshold $c$

````python
# Solution goes here
def train_fld(X, y):
    """Training the Fisher's Linear Discriminant model with input data X and target y.
    input - X, y
    output - the parameter vector w and decision threshold c."""
    
    # divide X into X1 and X2 according to which class x_n is in
    X1 = (X[np.flatnonzero(y > 0)])
    X2 = (X[np.flatnonzero(y < 0)])
    
    # compute the mean of data points in each class
    m1 = np.mean(X1,axis = 0)
    m2 = np.mean(X2,axis = 0)
    
    # compute within-class covariance matrix SW
    SW = np.dot((X1 - m1).T, (X1 - m1)) + np.dot((X2 - m2).T, (X2 - m2))
    
    # compute w using SW and m1, m2
    w = np.dot(np.linalg.inv(SW),(m2 - m1))
    
    # set the threshold c to 0
    # c = (np.dot(m1, w) * num_neg + np.dot(m2, w) * num_pos)/(num_pos + num_neg)
    c = 0
    
    #plot the project of X1 and X2 onto w
    ax = fig.add_subplot(121)    
    ax.plot(np.dot(X1, w), [0]*X1.shape[0], "bo", label="above 38")
    ax.plot(np.dot(X2, w), [0]*X2.shape[0], "go", label="below 38")
    ax.hist([np.dot(X1, w), np.dot(X2, w)])
    ax.set_title("Fisher's Linear Dis of training set C1 and C2")
    ax.legend()
    
    return w,c

def predict_fld(w, c, X):
    """ Use the parameter w, threshold c to predict class of input data X
    input - w, c, X
    output - a vector pred which is the prediction
    """
    
    pred = np.ones(len(X))
    
    #When w.T x > c the input data point is predicted to be in C1, 
    #otherwise the input data point is predicted to be in C2
    neg = np.dot(X,w) > c
    pred[neg] = -1
    
    return pred
````

````python
def confusionMatrix(labels_test, labels_predicted):
    """Compute the matrix of predictions versus labels"""
    if len(labels_test) != len(labels_predicted):
        return 0
    TP = 0; FP = 0; TN = 0; FN = 0
    for i in range(0, len(labels_test)):
        if labels_test[i] == 0 or labels_predicted[i] == 0:
            return 0
        if labels_test[i] > 0:
            if labels_predicted[i] > 0: TP += 1
            else: FN +=1
        else:
            if labels_predicted[i] > 0: FP += 1
            else: TN += 1
    return (TP, TN, FP, FN)

def accuracy(output, labels_test):
    """How many correct predictions?"""
    TP, TN, FP, FN = confusionMatrix(labels_test, np.sign(output))
    return float(TP + TN) / (TP + TN + FP + FN)
````

````python
fig = plt.figure(figsize=(11,5))

# train and predict
w,c = train_fld(X, y)
pred = predict_fld(w, c, X)
print(accuracy(pred, y))

# plot the right and wrong predictions
X1 = (X[np.flatnonzero(y > 0)])
X2 = (X[np.flatnonzero(y < 0)])

pred_X1 = predict_fld(w,c,X1)
pred_X2 = predict_fld(w,c,X2)

# plot the true_positive, false_positive, true_negative, false_negative results respectively
ax = fig.add_subplot(122)  
ax.plot(np.dot(X1, w)[pred_X1 > 0], pred_X1[pred_X1 > 0], "bo", label = "prediction age >= 38, correct")
ax.plot(np.dot(X2, w)[pred_X2 > 0], pred_X2[pred_X2 > 0], "rv", label = "prediction age >= 38, wrong")
ax.plot(np.dot(X1, w)[pred_X1 < 0], pred_X1[pred_X1 < 0], "ro", label = "prediction age < 38, wrong")
ax.plot(np.dot(X2, w)[pred_X2 < 0], pred_X2[pred_X2 < 0], "bv", label = "prediction age < 38, correct")
ax.set_title("Pridiction result of Fisher's Linear Dis")
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

````

####1. The Gaussian kernel
$$
    k(\mathbf x, \mathbf z) = exp \{- \dfrac{\Norm{\textbf x - \textbf z} ^ 2}{2 h^2}\} \tag{9}
$$

where,

* $h$ - the Gaussian kernel parameter controlling trade-off between sensitivity to noise and over-smoothing

* $\textbf x$, $\textbf z$ - two input data points

#### The matrix form of Gaussian kernel

Let $\mathbf{K}$ be the kernel matrix, $\mathbf{X}$ be the input matrix

According to the definition (9),

\begin{align*}
    \mathbf{K}_{ij} &= k(\mathbf{x}_i, \mathbf{x}_j) \\
    &= exp \{- \dfrac{\Norm{\textbf {x}_i - \textbf {x}_j} ^ 2}{2 h^2}\} \\
    &= exp \{- \dfrac{\mathbf{x}_i \mathbf{x}_i^T - 2 \mathbf{x}_i \mathbf{x}_j^T + \mathbf{x}_j \mathbf{x}_j^T}{2 h^2}\}
\end{align*}

Define a matrix $\mathbf{\hat{K}}$, such that 
$$
    \mathbf{\hat{K}}_{ij} = \mathbf{x}_i \mathbf{x}_i^T - 2 \mathbf{x}_i \mathbf{x}_j^T + \mathbf{x}_j \mathbf{x}_j^T \tag{10}
$$

we can compute $\mathbf K$ from $\mathbf{\hat{K}}$ by element-wise operation:

$$
    \mathbf K_{ij} = exp \{- \dfrac{\mathbf{\hat{K}}_{ij}}{2 h^2}\} \tag{11}
$$

Define $\mathbf{X}^2$ as a matrix such taht $\mathbf{X}^2_{ij} = (\mathbf{X}_{ij})^2$, $\onevec$ is a column vector of ones, we have

$$
    \mathbf{X}^2 \onevec
$$

to be a column vector where $(\mathbf{X}^2 \onevec) _i = \mathbf{x}_i \mathbf{x}_i^T$, and

$$
    \onevec^T \mathbf{X}^2 \onevec =  \left[ \begin{array}{ccc}
    \mathbf{x}_1 \mathbf{x}_1^T & \mathbf{x}_1 \mathbf{x}_1^T & ... &\mathbf{x}_1 \mathbf{x}_1^T \\
    \mathbf{x}_2 \mathbf{x}_2^T & \mathbf{x}_2 \mathbf{x}_2^T & ... &\mathbf{x}_2 \mathbf{x}_2^T \\
    ... & ... & ... & ...\\
    \mathbf{x}_N \mathbf{x}_N^T & \mathbf{x}_N \mathbf{x}_N^T & ... &\mathbf{x}_N \mathbf{x}_N^T \end{array} \right] \tag{12}
$$

and the transpose 
$$
    (\onevec^T \mathbf{X}^2 \onevec)^T = \onevec^T {\mathbf{X}^T}^2 \onevec \tag{13}
$$

We have 

$$
    \mathbf X \mathbf X^T = \left[ \begin{array}{ccc}
    \mathbf{x}_1^T \mathbf{x}_1 & \mathbf{x}_1^T \mathbf{x}_2 & ... &\mathbf{x}_1^T \mathbf{x}_N \\
    \mathbf{x}_2^T \mathbf{x}_1 & \mathbf{x}_2^T \mathbf{x}_2 & ... &\mathbf{x}_2^T \mathbf{x}_N \\

    \mathbf{x}_N^T \mathbf{x}_1 & \mathbf{x}_N^T \mathbf{x}_2 & ... &\mathbf{x}_N^T \mathbf{x}_N \end{array} \right] \tag{14}
$$

Because $\mathbf{x}_i \mathbf{x}_j^T$ is a number, the transpose of a number is itself, we have $\mathbf{x}_i \mathbf{x}_j^T = \mathbf{x}_j \mathbf{x}_i^T $

Comparing (12), (13) and (14) to (10), we have

$$
    \hat{\mathbf{K}} = \onevec^T \mathbf{X}^2 \onevec - 2 \mathbf X \mathbf X^T + \onevec^T {\mathbf{X}^T}^2 \onevec \tag{15}
$$

Substitute (15) into (11), we can compute the kernel matrix $\mathbf{K}$

where,

* $h$ - the Gaussian kernel parameter controlling trade-off between sensitivity to noise and over-smoothing

* $\textbf X$ - input data matrix. $\mathbf X \in \RR^{N \times D}$, where N is the total number of input data, and D is the dimension of the input data 

####2. The inhomogenous polynomial kernel
$$
    k(\mathbf x, \mathbf z) = (\mathbf x ^ T \mathbf z + c) ^ d \tag{16}
$$

where,

* $\mathbf x, \mathbf z$ - two input data points

* c - a non-negative constant, controlling the relative weight of different order terms

* d - the order of the polynomial kernel

#### The matrix form of inhomogenous polynomial kernel

Let $\mathbf{K}$ be the kernel matrix, $\mathbf{X}$ be the input matrix

According to the definition (16),

\begin{align*}
    \mathbf{K}_{ij} &= k(\mathbf{x}_i, \mathbf{x}_j) \\
    &= (\mathbf {x}_i ^ T \mathbf {x}_j + c) ^ d\\
\end{align*}

Define a matrix $\mathbf{\hat{K}}$, such that 
$$
    \mathbf{\hat{K}}_{ij} = \mathbf {x}_i ^ T \mathbf {x}_j \tag{17}
$$

we can compute $\mathbf K$ from $\mathbf{\hat{K}}$ by element-wise operation:

$$
    \mathbf K_{ij} = (\mathbf{\hat{K}}_{ij} + c)^d \tag{18}
$$

As shown in (14),

$$
    \mathbf{\hat{K}} = \mathbf{X} \mathbf{X}^T \tag{19}
$$

Substitute (19) into (18), we can compute the kernel matrix $\mathbf{K}$

where,

* c - a non-negative constant, controlling the relative weight of different order terms

* d - the order of the polynomial kernel

* $\textbf X$ - input data matrix. $\mathbf X \in \RR^{N \times D}$, where N is the total number of input data, and D is the dimension of the input data 


Implement two functions corresponding to the kernels above *(2 points)*.

````python
# Solution goes here
def Gaussian_kernel(x, z, sigma):
    """Using the Gaussian kernel to compute the distance between two input data points
    input - two data points x, z, the Gaussian parameter sigma
    output - the distance
    """
    return np.exp(-1 * (np.linalg.norm(x - z) ** 2) /(2 * (sigma ** 2)))

def Gaussian_kernel_matrix(X, sigma):
    """Using the Gaussian kernel to compute the kernel matrix for a input data matrix
    input - inpu datat matrix X, the Gaussian parameter sigma
    output - the kernel matrix K
    """
    X2 = np.sum(X ** 2, 1)
    K0 = X2[np.newaxis, :] + X2[:, np.newaxis] - 2 * np.dot(X, X.T)
    K = np.exp(-1 * K0 / (2 * (sigma ** 2))) 
    return K

def polynomial_kernel(x, z, c, d):
    """Using the inhomogenous polynomial kernel to compute the distance between two input data points
    input - two data points x, z, the bias term c, the dimension d
    output - the distance
    """
    return (np.dot(x.T, z) + c) ** d

def polynomial_kernel_matrix(X, c, d):
    """Using the inhomogenous polynomial kernel to compute the kernel matrix for a input data matrix
    input - input data matrix X, the bias term c, the dimension d
    output - the kernel matrix K
    """
    K0 = np.dot(X, X.T)
    K = (K0 + c) ** d
    return K
````

## (4 points) Normalisation in using kernels

You have seen the importance of normalisation in the tutorials.

In the following, we use the fact that kernels ($k(\cdot, \cdot)$) are inner products in a feature space with feature mapping $\phi(\cdot)$:
$$k(x,y) = \dotprod{\phi(x)}{\phi(y)}$$

### Centering

Centering causes the mean of the data set to be the zero vector in feature space. The following is a derivation for doing the centering directly using kernels.

$$
\mu = \frac{1}{n}\sum_{i=1}^n \phi(x_i) \tag{20}
$$
then
$$
\hat{\phi}(x) = \phi(x) - \mu. \tag{21}
$$
Hence
\begin{align*}
\hat{k}(x,y) &= \dotprod{\hat{\phi}(x)}{\hat{\phi}(y)} \\
    &= \dotprod{\phi(x) - \mu}{\phi(y) - \mu} \tag{22}
\end{align*}

Justify and explain the above steps.

### Solution description

In centering, we first compute the mean of the data in feature space by 
$$
    \mu = \frac{1}{n}\sum_{i=1}^n \phi(x_i) \tag{20}
$$

Then, we subtract the mean $\mu$ from each data point in feature space $\phi(x_i)$ The result is $\hat{\phi}(x_i)$, a shift of the data point in feature space 

$$
    \hat{\phi}(x) = \phi(x) - \mu \tag{21}
$$

#### We need to prove the mean of the data set in feature space is zero vector
\begin{align*}
    \hat{\mu} = \frac{1}{n} \sum_{i=1}^n \hat{\phi}(x_i) &= \frac{1}{n} \sum_{i=1}^n ( \phi(x_i) - \mu )\\
    &= \frac{1}{n} \sum_{i=1}^n \phi(x_i) - \frac{1}{n} \sum_{i=1}^n \mu \\
    &= \frac{1}{n} \sum_{i=1}^n \phi(x_i) - \mu
\end{align*}
as $\mu = \frac{1}{n}\sum_{i=1}^n \phi(x_i)$ given by (20), we have

$$
    \hat{\mu} = \frac{1}{n} \sum_{i=1}^n \phi(x_i) - \mu = \mathbf 0
$$
Hence, the mean of the data set in feature space is zero vector as required

#### Next, we prove $\hat{k}(x,y)$ is a valid kernel

$\hat{k}(x,y)$ is defined by

\begin{align*}
    \hat{k}(x,y) &= \dotprod{\hat{\phi}(x)}{\hat{\phi}(y)} \\
    &= \dotprod{\phi(x) - \mu}{\phi(y) - \mu} \tag{22}
\end{align*}

In (22), the first equation is the definition of a kernel matrix, and the second equation is got by substituting (21)

By the definition, to prove $\hat{k}(x,y)$ is a valid kernel, is equivalent to prove $\hat{K}$ is a positive semi-definite matrix, where $\hat{K}$ is the kernel matrix of $\hat{k}(x,y)$, given by

$$
    \hat{K}_{ij} = \hat{k}(x_i,x_j)
$$

According to (22), we have

\begin{align*}
    \hat{K}_{ij} &= \dotprod{\phi(x_i) - \mu}{\phi(x_j) - \mu} \\
    &= \dotprod{\phi(x_i)}{\phi(x_j)} - \dotprod{\phi(x_i)}{\mu} - \dotprod{\phi(x_j)}{\mu} + \dotprod{\mu}{\mu} \\
    &= K_{ij} - \frac{1}{n} \sum_{i=1}^n K_{ij} - \frac{1}{n} \sum_{j=1}^n K_{ij} + \frac{1}{n^2} \sum_{i,j=1}^n K_{ij} \tag{23}
\end{align*}

where $K$ is the kernel matrix of a valid kernel $k(x,y)$, and $K_{ij} = k(x_i,x_j)$

Let $ \Phi = [\phi(x_1) \phi(x_2) ... \phi(x_i)]^T $ ,  and $ \bar{\Phi} = [\mu \mu ... \mu]^T $

Compare to (23), we can see that 
$$
    \hat{K} = (\Phi - \bar{\Phi})(\Phi - \bar{\Phi})^T \tag{24}
$$

which shows that $\hat{K}$ is positive semi-definite as required

#### In (23), we can observe that centering is done directly using kernels, 
because all data points appears only in the form of elements in the kernel matrices, without explicitly referring to the features of a single data point.


### Unit diagonal

It is often convenient to have all the examples to be represented by vectors of the same length. This implies that the diagonal of the kernel matrix (the squared length) is the same for all examples. We arbitrarily (without loss of generality) set this length to 1.
\begin{align*}
\hat{k}(x,y) &= \dotprod{\frac{\phi(x)}{\|\phi(x)\|}}{\frac{\phi(y)}{\|\phi(y)\|}} \tag{25}\\
    &= \frac{1}{\|\phi(x)\|\|\phi(y)\|}\dotprod{\phi(x)}{\phi(y)}\\
    &= \frac{1}{\|\phi(x)\|\|\phi(y)\|} k(x,y) \tag{28}
\end{align*}

Normalizing the kernel matrix such that it has one along the diagonal is sometimes called trace normalisation or spherical normalisation.

Justify and explain each step of the derivation above.


### Solution description

In trace normalization, 

#### we present all feature vectors in the same length. Without loss of generality, we set the length to 1

$$
    \hat{k}(x,y) = \dotprod{\frac{\phi(x)}{\|\phi(x)\|}}{\frac{\phi(y)}{\|\phi(y)\|}} \tag{25}
$$

in which, the lengh of $\frac{\phi(x)}{\|\phi(x)\|}$ is presented as

\begin{align*}
    \|\frac{\phi(x)}{\|\phi(x)\|}\| &= \sqrt {\sum_{i=1}^{D} \frac{\phi_i(x)}{\|\phi(x)\|}^2} \\
    &= \frac{1}{\|\phi(x)\|} \sqrt{ \sum_{i=1}^{D} \phi_i(x)^2} \tag{26}
\end{align*}

where $D$ is the dimension of the feature vector, and $\frac{1}{\|\phi(x)\|}$ is the inverse of the length of $\phi(x)$, which is a positive number as shown below,

$$
    \|\phi(x)\| = \sqrt{\sum_{i=1}^{D} \phi_i(x)^2} \tag{27}
$$

when $\|\phi(x)\| \neq 0$, $\|\phi(x)\|$ is a positive number

Substitute (27) into (26), we have

\begin{align*}
    \|\frac{\phi(x)}{\|\phi(x)\|}\| &= \frac{1}{\sqrt{\sum_{i=1}^{D} \phi_i(x)^2}} \sqrt{ \sum_{i=1}^{D} \phi_i(x)^2} \\
    &= 1
\end{align*}

which means the length of $\frac{\phi(x)}{\|\phi(x)\|}$ equals to 1. Without loss of generality, the length of $\frac{\phi(x_i)}{\|\phi(x_i)\|}$ for all $x_i$ equals to 1, which proves all feature vectors are presented in the same length, 1, as required.

#### We will show $\hat{k}(x,y)$ can be presented in the form of $k(x,y)$ scaled by some multiplier

Starting from (25),

\begin{align*}
    \hat{k}(x,y) &= \dotprod{\frac{\phi(x)}{\|\phi(x)\|}}{\frac{\phi(y)}{\|\phi(y)\|}} \\
    &= \sum_{i=1}^{D} \frac{\phi_i(x) \phi_i(y)}{\|\phi(x)\| \|\phi(y)\|}
\end{align*}

as shown in (27), $\|\phi(x)\| \|\phi(y)\|$ is a positive nmuber. Take it out of the sum, we have

\begin{align*}
    \hat{k}(x,y) &= \frac{1}{\|\phi(x)\| \|\phi(y)\|} \sum_{i=1}^{D} \phi_i(x) \phi_i(y) \\
    &= \frac{1}{\|\phi(x)\| \|\phi(y)\|} \dotprod{\phi(x)}{\phi(y)} 
\end{align*}

Notice that $\dotprod{\phi(x)}{\phi(y)} = k(x, y)$, we have

$$
    \hat{k}(x,y) = \frac{1}{\|\phi(x)\| \|\phi(y)\|} k(x,y) \tag{28}
$$

as required.

#### Prove $\hat{k}(x,y)$ is a valid kernel

By the definition, to prove $\hat{k}(x,y)$ is a valid kernel, is equivalent to prove $\hat{K}$ is a positive semi-definite matrix, where $\hat{K}$ is the kernel matrix of $\hat{k}(x,y)$, given by

$$
    \hat{K}_{ij} = \hat{k}(x_i,x_j)
$$

According to (28), we have

\begin{align*}
    \hat{K}_{ij} &= \frac{1}{\|\phi(x)\| \|\phi(y)\|} k(x,y)\\
    &= \frac{1} {\sqrt {K_{ii} K_{jj}}} K_{ij} \tag{29}
\end{align*}

where $K$ is the kernel matrix of a valid kernel $k(x,y)$, and $K_{ij} = k(x_i,x_j)$

and $K^2$ is an element-wise squared matrix of $K$ given by,

$$
    K^2_{ij} = (K_{ij})^2
$$

From (29), we can write $\hat{K}$ as

$$
    \hat{K} = \tilde{K} \circ K
$$

where

$$
    \tilde{K}_{ij} = \frac{1} {\sqrt {K_{ii} K_{jj}}}
$$

According to [Schur Product Theorem](http://en.wikipedia.org/wiki/Schur_product_theorem), to prove $\hat{K}$ is positive semi-definite, we only $\tilde{K}$ and $K$ are positive semidefinite

By definition of a valid kernel $k(x,y)$, we know $K$ is positive semidefinite

Next, we prove $\tilde{K}$ is positive semidefinite

Define $\dot{K}$, $\dot{k}$ as

$$
    \dot{K}_{ij} = K_{ii} K_{jj} \\
    \dot{\mathbf{k}}_i = K_{ii}
$$

we have

$$
    \dot{K} = \dot{\mathbf k} \dot{\mathbf k}^T
$$

is a positive semidefinite matrix

Consider the eigenvalue decomposition of $\dot K = U^T \Lambda U$, when $K_{ii} \neq 0$, we have

$$
    \tilde K = U^T \sqrt {\Lambda ^ {-1}} U \tag{30}
$$

Where $(\sqrt {\Lambda ^ {-1}})_{ij} = \sqrt {(\Lambda_{ij}) ^ {-1}}$

(30) gives that $\tilde K$ is also a positive semidefinite matrix, as required.

#### Prove the diagonal of $\hat K$ equals 1, which means

\begin{align*}
    \hat K_{ii} &= \hat k(x_i, x_i) \\
    &= 1
\end{align*}

for all i

Substitute $x = x_i$, $y = x_i$ into (25) we have 

\begin{align*}
    \hat{k}(x,y) &= \dotprod{\frac{\phi(x_i)}{\|\phi(x_i)\|}}{\frac{\phi(x_i)}{\|\phi(x_i)\|}} \\
    &= \frac{\sum_{i = 1}^{D} \phi_i(x_i)^2}{\sqrt{\sum_{i = 1}^{D} \phi_i(x_i)^2} \sqrt{\sum_{i = 1}^{D} \phi_i(x_i)^2}}\\
    &= 1
\end{align*}

as required

#### In (29), we can observe that centering is done directly using kernels, 
because all data points appears only in the form of elements in the kernel matrices, without explicitly referring to the features of a single data point.


## (3 points) Kernelising Fisher's Discriminant

### Definition
Consider a binary classification task.
Recall from the lecture that Fisher's criterion is given by:
$$
J(w) = \frac{(m_2 - m_1)^2}{s_1^2 + s_2^2}
$$
where $m_1$ and $m_2$ are the means of class $\mathcal{C}_1$ and $\mathcal{C}_2$ respectively, and $s_1^2$ and $s_2^2$ are the corresponding within class variances.

Write down the definition of $m_1, m_2, s_1, s_2$ in terms of the examples $\mathbf{x}$ and the labels $y$. Please define all your symbols carefully, in particular $\mathbf{w}$.

### Solution
$\mathbf{m}_1$ and $\mathbf{m}_2$ are the means of class $\mathcal{C}_1$ and $\mathcal{C}_2$ given by
$$ 
    \mathbf{m}_1 = \frac{1}{N_1} \sum_{n \in \mathcal{C}_1} \mathbf {x}_n \tag{31}     
$$
$$
    \mathbf{m}_2 = \frac{1}{N_2} \sum_{n \in \mathcal{C}_2} \mathbf {x}_n \tag{32}
$$
where,
* $N_1$ is the number of input data in $\mathcal{C}_1$
* $N_2$ is the number of input data in $\mathcal{C}_2$

The projection of $\mathbf{m_1}$ and $\mathbf{m_2}$ onto $\mathbf{w}$, are denoted by $\mathbf{p_1}$ and $\mathbf{p_2}$
$$ 
    \mathbf{p_1} = \frac{\mathbf{w} \mathbf{w}^T}{\mathbf{w}^T \mathbf{w}} \mathbf{m_1} \\
    \mathbf{p_2} = \frac{\mathbf{w} \mathbf{w}^T}{\mathbf{w}^T \mathbf{w}} \mathbf{m_2}
$$

We have
$$
    \mathbf{p_1} - \mathbf{p_2} = \frac{\mathbf{w}}{\mathbf{w}^T \mathbf{w}} (\mathbf{w}^T \mathbf{m}_1 - \mathbf{w}^T \mathbf{m}_2)
$$

in which $\frac{\mathbf{w}}{\mathbf{w}^T \mathbf{w}}$ is a vector in the direction of $\mathbf{w}$ and $(\mathbf{w}^T \mathbf{m}_1 - \mathbf{w}^T \mathbf{m}_2)$ is the scalar of the vector

#### To maximize $|\mathbf{p_1} - \mathbf{p_2}|$ is equivalent to maximize $|m_1 - m_2|$, given by
$$
    m_1 - m_2 = \mathbf{w}^T (\mathbf{m}_1 - \mathbf{m}_2) \tag{33}
$$
where
$$
    m_1 = \mathbf{w}^T \mathbf{m}_1 \\
    m_2 = \mathbf{w}^T \mathbf{m}_2
$$
$m_1$ and $m_2$ are the scalar of the projection of $\mathbf{m}_1$ and $\mathbf{m}_2$ onto $\mathbf{w}$

#### $s_1$ and $s_2$ are the within class variance of y's from $\mathcal{C}_1$ and $\mathcal{C}_2$ given by
$$ 
    s_1^2 = \sum_{n \in \mathcal{C}_1} (y_n - m_1)^2 \\
    s_2^2 = \sum_{n \in \mathcal{C}_2} (y_n - m_2)^2
$$

where $y_n = \mathbf{w}^T \mathbf{x}_n$

$s_1^2$ and $s_2^2$ denote the sum of squared differences, between (the scalar of) the projection onto $\mathbf{w}$ of inputdata, and the projection of their means onto $\mathbf{w}$ within a class, which, in other words, is the within class variance of projections

#### $\mathbf{w}$ is chosen by maximize the Fisher criterion
$$
    J(w) = \frac{(m_2 - m_1)^2}{s_1^2 + s_2^2}
$$
For the dataset of two classes to be as seperated as possible under a projection $\mathbf{w}$
* The bigger the difference of their means $(m_2 - m_1)^2$ the better
* The smaller the within class variance $(s_1^2 + s_2^2)$ the better

### Matrix form

Observe that you can express the sum of a set of vectors as a product of a vector of ones, $\onevec$ and a matrix containing the vectors. Define matrices $\mathbf{X}_1$ and $\mathbf{X}_2$ as the data corresponding to class $\mathcal{C}_1$ and $\mathcal{C}_2$ respectively. Please specify the dimensions carefully. Using these definitions, derive the expression for the numerator of $J(w)$, i.e. $(m_2 - m_1)^2$ where the data only appears in terms of the matrices $\mathbf{X}_1\mathbf{X}_1^T$, $\mathbf{X}_1\mathbf{X}_2^T$, $\mathbf{X}_2\mathbf{X}_1^T$ and $\mathbf{X}_2\mathbf{X}_2^T$. There could be other vectors, for example $\mathbf{w}$.

### Solution

According to (33)

\begin{align*}
    (m_2 - m_1)^2 &= (\mathbf{w}^T \mathbf{m}_1 - \mathbf{w}^T \mathbf{m}_2)^2  \tag{34}
\end{align*}

where $\mathbf{w}$, $\mathbf{m}_1$, $\mathbf{m}_2$ $\in \RR^{D \times 1}$, $D$ is the dimension of the input data

Let 

\begin{align*}
    \mathbf{w} &= \sum_{n = 1}^{N} \mathbf {\alpha}_n \mathbf{x}_n \\
    &= \sum_{n \in \mathcal{C_1}} \mathbf {\alpha}_n \mathbf{x}_n + \sum_{n \in \mathcal{C_2}} \mathbf {\alpha}_n \mathbf{x}_n\\
    &= \mathbf{X}_1^T \mathbf{\alpha}_1 + \mathbf{X}_2^T \mathbf{\alpha}_2 \tag{35}
\end{align*}

in which $\mathbf{X}_1$ and $\mathbf{X}_2$ as the data corresponding to class $\mathcal{C}_1$ and $\mathcal{C}_2$ respectively

we have $\mathbf{X}_1 \in \RR^{N_1 \times D}$, $\mathbf{X}_2 \in \RR^{N_2 \times D}$, $\mathbf \alpha \in \RR^{N \times 1}$, 
$\mathbf{\alpha}_1 \in \RR^{N_1 \times 1}$, $\mathbf{\alpha}_2 \in \RR^{N_2 \times 1}$, the result $\mathbf{w} \in \RR^{D \times 1}$

According to (31)
\begin{align*}
    \mathbf {m}_1 &= \frac{1}{N_1} \sum_{n \in \mathcal {C}_1} \mathbf {x}_n \\
    &= \frac{1}{N_1}  \mathbf {X_1}^T \onevec \tag{36}
\end{align*}

where  $\mathbf {X}_1 \in \RR^{N_1 \times D}$, $\onevec \in \RR^{N_1 \times 1}$,  $\mathbf {X}_1^T \onevec \in \RR^{D \times 1}$

Similarly, according to (32)
$$
    \mathbf {m}_2 = \frac{1}{N_2}  \mathbf {X_2}^T \onevec \tag{37}
$$

where  $\mathbf {X}_2 \in \RR^{N_2 \times D}$, $\onevec \in \RR^{N_2 \times 1}$,  $\mathbf {X}_2^T \onevec \in \RR^{D \times 1}$

Substitute (35), (36) and (37) into (34), we get

\begin{align*}
    (m_2 - m_1)^2 &= (\mathbf{w}^T \mathbf{m}_1 - \mathbf{w}^T \mathbf{m}_2)^2 \\
    &= ((\mathbf{X}_1^T \mathbf{\alpha}_1 + \mathbf{X}_2^T \mathbf{\alpha}_2)^T \frac{1}{N_1} \mathbf{X}_1^T \onevec - (\mathbf{X}_1^T \mathbf{\alpha}_1 + \mathbf{X}_2^T \mathbf{\alpha}_2)^T \frac{1}{N_2} \mathbf{X}_2^T \onevec)^2\\
    &= (\frac{1}{N_1}(\mathbf{\alpha}_1^T \mathbf{X}_1 \mathbf{X}_1^T \onevec + \mathbf{\alpha}_2^T \mathbf{X}_2 \mathbf{X}_1^T \onevec) - \frac{1}{N_2}(\mathbf{\alpha}_1^T \mathbf{X}_1 \mathbf{X}_2^T \onevec + \mathbf{\alpha}_2^T \mathbf{X}_2 \mathbf{X}_2^T \onevec))^2
\end{align*}

where $\mathbf{X}_1^T \mathbf{\alpha}_1 + \mathbf{X}_2^T \mathbf{\alpha}_2 = \mathbf{w}$, and the data only appears in terms of the matrices $\mathbf{X}_1\mathbf{X}_1^T$, $\mathbf{X}_1\mathbf{X}_2^T$, $\mathbf{X}_2\mathbf{X}_1^T$ and $\mathbf{X}_2\mathbf{X}_2^T$

### Kernel form

The above matrix forms $\mathbf{X}_1\mathbf{X}_1^T$, $\mathbf{X}_1\mathbf{X}_2^T$, $\mathbf{X}_2\mathbf{X}_1^T$ and $\mathbf{X}_2\mathbf{X}_2^T$ can be considered to be the special case of the linear kernel, i.e. $\phi(x) = x$. Observe that the numerator of Fisher's criterion can be expressed purely as inner products between examples. It turns out that the denominator can also be expressed purely as inner products.


Implement Kernel Fisher's Discriminant such that it can take a general kernel function. This is the function of the form:
$$
J(\alpha) = \frac{\alpha M \alpha}{\alpha N \alpha}
$$
where $M$ and $N$ are defined on the following [Wikipedia entry](http://en.wikipedia.org/wiki/Kernel_Fisher_discriminant_analysis).

````python
# Solution goes here

def centering(K):
    """ Normalize kernel matrix using centering
    input - kernel matrix K
    output - centered kernel matrix, with mean(K_centered) = 0
             centering parameters: row_mean, col_mean, K_mean 
    """
    row_mean = np.mean(K, axis = 0)
    col_mean = np.mean(K, axis = 1)
    K_mean = np.mean(K)
    K_centered = K - row_mean - col_mean + K_mean
    return row_mean, col_mean, K_mean, K_centered

def unit_diagonal(K):
    """ Normalize kernel matrix with unit feature vector
    input - kernel matrix K
    output - normalized kernel matrix, with diagonal(K_normalized) = 1
             unit diagonal parameter: D
    """
    D = np.diag(1.0 / np.sqrt(np.diag(K)))
    K_normalized = np.dot(np.dot(D, K), D)
    return D, K_normalized 

def kernel_fisher_discriminant(K, y):
    """ Kernel Fisher's discriminant analysis. Using the training kernel matrix and target to fit the parameter alpha
    input - kernel matrix K, target y
    output - parameter alpha, 
             normalized kernel matrix K, 
             normalization parameters: row_mean, col_mean, K_mean, D
    """
    
    # Normalize K with one of the two normalization methods: centering or unit diagonal
    row_mean, col_mean, K_mean, K_centered = centering(K)
    D, K_normalized = unit_diagonal(K)
    #K = K_normalized
    K = K_centered
    
    # Compute M1 and M2
    K1 = (K[np.flatnonzero(y > 0)]).T
    K2 = (K[np.flatnonzero(y < 0)]).T
    M1 = np.mean(K1, axis = 1)
    M2 = np.mean(K2, axis = 1)
    
    # Compute N
    I1 = np.identity(K1.shape[1])
    I2 = np.identity(K2.shape[1])
    inv_N1 = np.ones((K1.shape[1], K1.shape[1])) * 1.0 / K1.shape[1]
    inv_N2 = np.ones((K2.shape[1], K2.shape[1])) * 1.0 / K2.shape[1]
    N = np.dot(np.dot(K1, I1 - inv_N1), K1.T) + np.dot(np.dot(K2, I2- inv_N2), K2.T)
    N = N +  np.identity(N.shape[1]) * 0.1
    
    # Comute alpha with M1, M2 and N 
    alpha = np.dot(np.linalg.inv(N), (M2 - M1))

    return row_mean, col_mean, K_mean, D, K, alpha
````
